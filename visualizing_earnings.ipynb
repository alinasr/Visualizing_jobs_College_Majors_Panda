# Visualizing Earnings Based On College Majors

In this project we aimed to analysis the jobs outcome for majors only by using plotting tools built into pandas just for practicing these tools

in the code blow we open the data set and show some rows from this data set

```python
import pandas as pd
import matplotlib as plt
%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv')

first_row = recent_grads.iloc[:1]
print(first_row)
recent_grads.head()
```

here is some information about the data set by the code blow

```python
recent_grads.info()
recent_grads.describe()
```

## Cleaning missing data

We should delete the rows with missing values because they will cause matplotlib to throw errors.

In the code blow we dropped the rows with missing values and if we look at the number of rows before and after dropping we will see there was only one row with missing value


```python
raw_data_count, _ = recent_grads.shape
print(raw_data_count)

recent_grads = recent_grads.dropna()

cleaned_data_count, _ = recent_grads.shape
print(cleaned_data_count)
```

## Scatter Plots

Here we are going to plot some scatter plots for some columns of our data set to for exploring some information


```python
recent_grads.plot(x = 'Total', y = 'Median', kind = 'scatter',
                 title='Sample_size vs. Median', figsize=(5,5))
```

```python
recent_grads.plot(x = 'Sample_size', y = 'Median', kind = 'scatter',
                 title='Sample_size vs. Median', figsize=(5,5))
```

As we can see in the plot above we can conclude that popular majors have more median income in average

```python
recent_grads.plot(x = 'Sample_size', y = 'Unemployment_rate', kind = 'scatter',
                 title='Sample_size vs. Unemployment_rate', figsize=(5,5))
```

Unemployment rate for less popular majors is vary a low as we can see but for popular the range vary less.

```python
recent_grads.plot(x = 'Full_time', y = 'Median', kind = 'scatter',
                 title='Full_time vs. Median', figsize=(5,5))
```

As we can see full time jobs have more certain amount of earning

```python
recent_grads.plot(x = 'ShareWomen', y = 'Unemployment_rate', kind = 'scatter',
                 title='ShareWomen vs. Unemployment_rate', figsize=(5,5))
```

as you can see there is no correlation between Unemployment rate and sharewomane

```python
recent_grads.plot(x = 'Men', y = 'Median', kind = 'scatter',
                 title='Men vs. Median', figsize=(5,5))
```

```python
recent_grads.plot(x = 'Women', y = 'Median', kind = 'scatter',
                 title='Women vs. Median', figsize=(5,5))
```

Based on 2 previous plots we can see slightly majority female do not make more money

## Scatter matrix plot

A scatter matrix plot combines both scatter plots and histograms into one grid of plots and allows us to explore potential relationships and distributions simultaneously.

Here we are going to plot scatter plots in the code cells blow


```python
from pandas.plotting import scatter_matrix

scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,10))
```

We can see in a better view that there is slight correlation between sample size and medina income. it shows as the most popular jobs have more income in average. Also we can conclude most of the sample sizes are less than 500 and the most common median salary is 30000

```python
scatter_matrix(recent_grads[['Sample_size', 'Median','ShareWomen']], figsize=(10,10))
```

We can see only correlation between sharewoman and median income. It shows us in the majors with more woman the median income is lower in compare to majors with less woman witch there is high income. Also we can see that most of the majors have the woman share between 30% and 70%

```python
scatter_matrix(recent_grads[['Full_time', 'Median', "Sample_size"]], figsize=(10,10))
```

Also we can see in a better view the fulltime employees have more median salary in average. There is a good correlation between sample size and full time jobs. Also most of the majors have 0-4000 number of full time employees.

## Bar Plots

here we are going to draw bar plots for share woman for 10 first and 10 last majors to see the relation between share woman and high and low income majors. Also for the unemployment rate for 10 first and 10 last majors to see the relation between unemployment rate and high and low income majors

```python
recent_grads[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
recent_grads[163:].plot.bar(x='Major', y='ShareWomen', legend=False)
```

As we can see in the plot above because the data frame ordered based on median income from high to low. The plot shows us the women share increase in low income majors. The highest percentage for first 10 row is 50% but for last ten rows the lowest percentage is 60%.

```python
recent_grads[:10].plot.bar(x='Major', y='Unemployment_rate', legend=False)
recent_grads[163:].plot.bar(x='Major', y='Unemployment_rate', legend=False)
```

The plot in above is Unemployment_rate for 10 high income and 10 low income majors. The variation in Unemployment_rate in high income majors is higher in compare to low income major.
